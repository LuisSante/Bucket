```python
import cv2
import numpy as np
import matplotlib.pyplot as plot
```

```python
img_grayscale = cv2.imread('C:/Users/USUARIO/Desktop/opencv-python/images.jpg',0)
```

```python
def histogram_image(image):
    pixeles, frecuency = np.unique(image, return_counts=True)

    plot.figure(figsize=(12,12))
    plot.hist(x = image.ravel(), bins=256, fc = 'k', ec = 'k')
    plot.title('Histograma')
    plot.xlabel('Pixel')
    plot.ylabel('Frecuencia')   
    plot.show()

    return pixeles, frecuency
```

```python
def show_image(image):
    cv2.imshow('graycsale image',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.imwrite('grayscale.jpg',img_grayscale)
```

```python
def value_min(list_frec):
    minors = []
    arr = np.array(list_frec)
    min_ = min(arr)
    result = np.where(arr == min_)
    result = list(result)
    minors.append({"pixel":result[0],"intensidad":min_})
    return minors
```

```python
pixeles, frequency = histogram_image(img_grayscale)
minors = value_min(frequency)
print("Pixeles\n ",pixeles)
print("Intensidad\n ",frequency)
print(minors)
```

```python
show_image(img_grayscale)
```
