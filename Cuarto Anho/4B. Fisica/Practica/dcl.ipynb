```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def aceleracion_tension():
    m_1 = float(input("ingrese masa 1 "))
    m_2 = float(input("ingrese masa 2 "))
    a = 0
    g = 9.8
    if(m_1 > m_2):
        a = g*((m_1 - m_2)/(m_1 + m_2))

    elif(m_1 < m_2):
        a = g*((m_2 - m_1)/(m_1 + m_2))

    elif(m_1 == m_2): 
        a = 0

    tension = (2*g*m_1*m_2)/(m_1+m_2)

    return a, tension
```

```python
def relacion_fuerza_aceleracion():
    a = 0
    F = 0  
    aceleracion = []
    fuerza = []
    count = 0

    while( count < 5):
        m = int(input("ingrese masa "))
        t = int(input("ingrese tiempo "))
        d = int(input("ingrese distancia " ))
        v0 = int(input("ingrese velocidad inicial "))
        vf = int(input("ingrese velocidad final "))
        
        if (t >= 0 and d >= 0 and v0 >= 0 and vf>=0 and m > 0):
            a = (vf-v0)/t
            F = m*a

            aceleracion.append(a)
            fuerza.append(F)
        count+=1

    plt.title("")
    plt.xlabel('aceleracion')
    plt.ylabel('Fuerza')
    plt.yticks(fuerza)
    plt.plot(aceleracion, fuerza, marker = 'o', c = 'g', label="procesos")
    plt.legend()
    plt.show()
```

```python
relacion_fuerza_aceleracion()
```

```python
aceleracion, tension = aceleracion_tension()
print(aceleracion, "m/s2")
print(tension, "N")
```
