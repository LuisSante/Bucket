```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = ["25%", "50%" , "100%"]
y = [11.414610385894775 , 24.082294464111328 , 43.90809082984924]

# Crear el gráfico
plt.plot(x, y)

# Personalizar el gráfico
plt.title('Big Data Experimentos')
plt.xlabel('Tamaño del dataset')
plt.ylabel('Tiempo')

# Mostrar el gráfico
plt.show()
```
